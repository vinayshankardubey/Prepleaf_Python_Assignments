### 4. Remove column Sr No, Remarks columns and print first 10 rows

```python
df.columns = df.columns.str.replace(' ', '')
df.drop(['SrNo'], axis=1, inplace=True)
```

### 5. Check for Null Values and remove all rows containing null values


```python
nan_rows = df[df.isna().any(axis=1)]
lst_row_nan = nan_rows.index.values.astype(int)
df.drop(lst_row_nan, axis=0, inplace=True)
```

### 10. Print Average AmountUSD for each InvestmentType
* Hint- use group by


```python
nd = df.groupby('InvestmentnType')
for type,type_d in nd:
    print(f"{type} : {type_d['AmountUSD'].astype(int).mean()}")
```
